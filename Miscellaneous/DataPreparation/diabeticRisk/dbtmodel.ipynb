```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.naive_bayes import MultinomialNB
```

```python
# Loading the dataset
dbt_data = pd.read_csv('./diabetes_binary.csv')
```

```python
# Create scaler
scaler = MinMaxScaler(feature_range=(0, 1))
```

```python
dbt_data['Diabetes_01']
```

```python
# categorical
HighBP = pd.Categorical(dbt_data['HighBP'])
HighChol = pd.Categorical(dbt_data['HighChol'])
CholCheck = pd.Categorical(dbt_data['CholCheck'])
Smoker = pd.Categorical(dbt_data['Smoker'])
Stroke = pd.Categorical(dbt_data['Stroke'])
HeartDiseaseorAttack = pd.Categorical(dbt_data['HeartDiseaseorAttack'])
PhysActivity = pd.Categorical(dbt_data['PhysActivity'])
Fruits = pd.Categorical(dbt_data['Fruits'])
Veggies = pd.Categorical(dbt_data['Veggies'])
HvyAlcoholConsump = pd.Categorical(dbt_data['HvyAlcoholConsump'])
AnyHealthcare = pd.Categorical(dbt_data['AnyHealthcare'])
NoDocbcCost = pd.Categorical(dbt_data['NoDocbcCost'])
# GenHlth = pd.Categorical(dbt_data['GenHlth'])
DiffWalk = pd.Categorical(dbt_data['DiffWalk'])
Sex = pd.Categorical(dbt_data['Sex'])
# Age = pd.Categorical(dbt_data['Age'])
# Education = pd.Categorical(dbt_data['Education'])
# Income = pd.Categorical(dbt_data['Income'])
Diabetes_01 = pd.Categorical(dbt_data['Diabetes_01'])

#scaling
dbt_data[['BMI','MentHlth','PhysHlth','Age','GenHlth','Education','Income']] = scaler.fit_transform(dbt_data[['BMI','MentHlth','PhysHlth','Age','GenHlth','Education','Income']])

dbt_data['HighBP'] = HighBP
dbt_data['HighChol'] = HighChol
dbt_data['CholCheck'] = CholCheck
dbt_data['Smoker'] = Smoker
dbt_data['Stroke'] = Stroke
dbt_data['HeartDiseaseorAttack'] = HeartDiseaseorAttack
dbt_data['PhysActivity'] = PhysActivity
dbt_data['Fruits'] = Fruits
dbt_data['Veggies'] = Veggies
dbt_data['HvyAlcoholConsump'] = HvyAlcoholConsump
dbt_data['AnyHealthcare'] = AnyHealthcare
dbt_data['NoDocbcCost'] = NoDocbcCost
# dbt_data['GenHlth'] = GenHlth
dbt_data['DiffWalk'] = DiffWalk
dbt_data['Sex'] = Sex
# dbt_data['Age'] = Age
# dbt_data['Education'] = Education
# dbt_data['Income'] = Income
dbt_data['Diabetes_01'] = Diabetes_01

# X = dbt_data.drop(['Diabetes_01'], axis=1)
X = dbt_data.drop(['Diabetes_01','CholCheck','AnyHealthcare','NoDocbcCost',
                   'GenHlth','Education','Income'], axis=1)
X
```

```python
Y_classification = dbt_data.Diabetes_01
```

```python
# Split train/test 
X_train, X_test, y_train, y_test = train_test_split(X, Y_classification, test_size=1 / 5, random_state=1,
                                                    stratify=Y_classification)
# print(X_train.shape)
# print(X_test.shape)
```

```python
NaiveBayesModel = MultinomialNB()
NaiveBayesModel.fit(X_train, y_train)

# print('Score in training set: {:.3f}'.format(NaiveBayesModel.score(X_train, y_train)))
# print('Score in test set: {:.3f}'.format(NaiveBayesModel.score(X_test,y_test)))
```

```python
def pred(data):
    prediction = NaiveBayesModel.predict(data)
    return prediction
```
