# The purpose of this code is to process the data from the raw data of year 2013 and combine with the processed data 2015. 
# Then generate two balance datasets, 1st balance dataset with Diabetes_01 is for binary classification purpose, 2nd balance dataset with Diabetes_012 is for multi-classification purpose.

```python
#imports
import os
import pandas as pd
import numpy as np
import random
random.seed(1)
```

```python
# read from raw data
dbt_data = pd.read_csv('./brfss2013.csv',encoding='ISO-8859-1')
```

```python
pd.set_option('display.max_columns', 500)
dbt_data.head()
```

```python
# select specific columns
brfss_df_selected = dbt_data[['diabete3',
                                         'X_rfhype5',  
                                         'toldhi2', 'X_cholchk', 
                                         'X_bmi5', 
                                         'smoke100', 
                                         'cvdstrk3', 'cvdinfr4', 
                                         'X_totinda', 
                                         'X_frtlt1', 'X_veglt1', 
                                         'X_rfdrhv4', 
                                         'hlthpln1', 'medcost', 
                                         'genhlth', 'menthlth', 'physhlth', 'diffwalk', 
                                         'sex', 'X_ageg5yr', 'educa', 'income2' ]]
```

```python
brfss_df_selected.shape
```

```python
#rename the selected features
brfss = brfss_df_selected.rename(columns = {'diabete3':'Diabetes_012', 
                                         'X_rfhype5':'HighBP',  
                                         'toldhi2':'HighChol', 'X_cholchk':'CholCheck', 
                                         'X_bmi5':'BMI', 
                                         'smoke100':'Smoker', 
                                         'cvdstrk3':'Stroke', 'cvdinfr4':'HeartDiseaseorAttack', 
                                         'X_totinda':'PhysActivity', 
                                         'X_frtlt1':'Fruits', 'X_veglt1':"Veggies", 
                                         'X_rfdrhv4':'HvyAlcoholConsump', 
                                         'hlthpln1':'AnyHealthcare', 'medcost':'NoDocbcCost', 
                                         'genhlth':'GenHlth', 'menthlth':'MentHlth', 'physhlth':'PhysHlth', 'diffwalk':'DiffWalk', 
                                         'sex':'Sex', 'X_ageg5yr':'Age', 'educa':'Education', 'income2':'Income' })
```

```python
brfss.head()
```

```python
# drop the incomplete data
brfss = brfss.dropna()
brfss.shape
```

```python
#remove data type 'Yes, but female told only during pregnancy'
brfss = brfss[brfss.Diabetes_012 != 'Yes, but female told only during pregnancy']

#change 'No' to 0, 'Yes' to 2, 'No, pre-diabetes or borderline diabetes' to 1
brfss['Diabetes_012'] = brfss['Diabetes_012'].replace({'No':0,'No, pre-diabetes or borderline diabetes':1, 'Yes':2})
brfss.Diabetes_012.unique()
```

```python
brfss.shape
```

```python
# this is just to find out the categorie types from the raw data of 'HighBP'
# HighBP = pd.Categorical(brfss['HighBP'])
# HighBP
```

```python
# HighBP
brfss['HighBP'] = brfss['HighBP'].replace({'No':0, 'Yes':1})
brfss.HighBP.unique()
brfss.shape
```

```python
#HighChol
# replace value from string to integer
brfss['HighChol'] = brfss['HighChol'].replace({'No':0, 'Yes':1})

#remove duplicate data
brfss.HighChol.unique()
```

```python
# CholCheck
brfss['CholCheck'] = brfss['CholCheck'].replace({'Did not have cholesterol checked in past 5 years':0,'Had cholesterol checked in past 5 years':1})
brfss.CholCheck.unique()
```

```python
#remove CholCheck = 0 but highchol = 1 - didn't do cholesterol check in 5 years but have high cholesterol,no that reliable
brfss = brfss[~((brfss['CholCheck'] == 0) & (brfss['HighChol'] != 0))]
```

```python
# BMI = pd.Categorical(brfss['BMI'])
# BMI
```

```python
# BMI
brfss['BMI'] = brfss['BMI'].div(100).round(0)
brfss.BMI.unique()
```

```python
# Smoker = pd.Categorical(brfss['Smoker'])
# Smoker
```

```python
# Smoker
brfss['Smoker'] = brfss['Smoker'].replace({'No':0, 'Yes':1})
brfss.Smoker.unique()
```

```python
# Stroke = pd.Categorical(brfss['Stroke'])
# Stroke
```

```python
# Stroke
brfss['Stroke'] = brfss['Stroke'].replace({'No':0, 'Yes':1})
brfss.Stroke.unique()
```

```python
# HeartDiseaseorAttack
brfss['HeartDiseaseorAttack'] = brfss['HeartDiseaseorAttack'].replace({'No':0, 'Yes':1})
brfss.HeartDiseaseorAttack.unique()
```

```python
# PhysActivity
brfss['PhysActivity'] = brfss['PhysActivity'].replace({'No physical activity or exercise in last 30 days':0, 'Had physical activity or exercise':1})
brfss.PhysActivity.unique()
```

```python
# Fruits
brfss['Fruits'] = brfss['Fruits'].replace({'Consumed fruit less than one time per day':0, 'Consumed fruit one or more times per day':1})
brfss.Fruits.unique()
```

```python
# Veggies
brfss['Veggies'] = brfss['Veggies'].replace({'Consumed vegetables less than one time per day':0, 'Consumed vegetables one or more times per day':1})
brfss.Veggies.unique()
```

```python
# HvyAlcoholConsump
brfss['HvyAlcoholConsump'] = brfss['HvyAlcoholConsump'].replace({'No':0, 'Yes':1})
brfss.HvyAlcoholConsump.unique()
```

```python
# AnyHealthcare
brfss['AnyHealthcare'] = brfss['AnyHealthcare'].replace({'No':0, 'Yes':1})
brfss.AnyHealthcare.unique()
```

```python
# NoDocbcCost
brfss['NoDocbcCost'] = brfss['NoDocbcCost'].replace({'No':0, 'Yes':1})
brfss.NoDocbcCost.unique()
```

```python
# GenHlth
brfss['GenHlth'] = brfss['GenHlth'].replace({'Excellent':1, 'Very good':2, 'Good':3, 'Fair':4,'Poor':5,})
brfss.GenHlth.unique()
```

```python
# MentHlth = pd.Categorical(brfss['MentHlth'])
# MentHlth
```

```python
# PhysHlth = pd.Categorical(brfss['PhysHlth'])
# PhysHlth
```

```python
# DiffWalk = pd.Categorical(brfss['DiffWalk'])
# DiffWalk
```

```python
# DiffWalk
brfss['DiffWalk'] = brfss['DiffWalk'].replace({'No':0, 'Yes':1})
brfss.DiffWalk.unique()
```

```python
# Sex
brfss['Sex'] = brfss['Sex'].replace({'Female':0, 'Male':1})
brfss.Sex.unique()
```

```python
# Age
brfss['Age'] = brfss['Age'].replace({'Age 18 to 24':1, 'Age 25 to 29':2, 'Age 30 to 34':3, 'Age 35 to 39':4, 'Age 40 to 44':5, 'Age 45 to 49':6, 'Age 50 to 54':7, 'Age 55 to 59':8, 'Age 60 to 64':9, 'Age 65 to 69':10, 'Age 70 to 74':11, 'Age 75 to 79':12, 'Age 80 or older':13})
brfss.Age.unique()
```

```python
# Education = pd.Categorical(brfss['Education'])
# Education
```

```python
# Education
brfss['Education'] = brfss['Education'].replace({'Never attended school or only kindergarten':1, 'Grades 1 through 8 (Elementary)':2, 'Grades 9 though 11 (Some high school)':3, 'Grade 12 or GED (High school graduate)':4, 'College 1 year to 3 years (Some college or technical school)':5, 'College 4 years or more (College graduate)':6})
brfss.Education.unique()
```

```python
# Income = pd.Categorical(brfss['Income'])
# Income
```

```python
# Income
brfss['Income'] = brfss['Income'].replace({'Less than $10,000':1, 'Less than $15,000':2, 'Less than $20,000':3, 'Less than $25,000':4, 'Less than $35,000':5, 'Less than $50,000':6, 'Less than $75,000':7, '$75,000 or more':8})
brfss.Income.unique()
```

```python
brfss.shape
```

```python
brfss.head()
```

```python
brfss['BMI'] = pd.to_numeric(brfss['BMI'], downcast="integer")
```

```python
brfss['MentHlth'] = pd.to_numeric(brfss['MentHlth'], downcast="integer")
```

```python
brfss['PhysHlth'] = pd.to_numeric(brfss['PhysHlth'], downcast="integer")
```

# crerate diabetes 01 balance dataset
# combine 0 - Non-diabetic and 1 - Pre-diabetic 

```python
brfss_binary = brfss1315
```

```python
# remove didn't do cholcheck in 5 years but have highchol
brfss_binary = brfss_binary[~((brfss_binary['CholCheck'] == 0) & (brfss_binary['HighChol'] != 0))]
brfss_binary.shape
```

```python
#combine non-diabetic & pre-diabetic as 0 , change diabetic from 2 to 1
brfss_binary['Diabetes_012'] = brfss_binary['Diabetes_012'].replace({1:0})
brfss_binary['Diabetes_012'] = brfss_binary['Diabetes_012'].replace({2:1})

brfss_binary = brfss_binary.rename(columns = {'Diabetes_012': 'Diabetes_01'})
brfss_binary.Diabetes_01.unique()
```

```python
brfss_binary.groupby(['Diabetes_01']).size()
```

```python
# Get the 1s
is1 = brfss_binary['Diabetes_01'] == 1
brfss_1 = brfss_binary[is1]

#Get the 0s
is0 = brfss_binary['Diabetes_01'] == 0
brfss_0 = brfss_binary[is0] 
```

```python
# Randomly pick up the same number of 1s from 0s
brfss_0_rand1 = brfss_0.take(np.random.permutation(len(brfss_0))[:77393])
```

```python
# combine into a balance dataset
brfss_5050 = brfss_0_rand1.append(brfss_1, ignore_index = True)
```

```python
brfss_5050.groupby(['Diabetes_01']).size()
```

```python
brfss_5050.tail()
```

```python
# Generate out the csv
brfss_5050.to_csv('diabetes_binary.csv', sep=",", index=False)
```

# create diabetes 012 balance dataset

```python
# remove didn't do cholcheck in 5 years but have highchol
brfss1315 = brfss1315[~((brfss1315['CholCheck'] == 0) & (brfss1315['HighChol'] != 0))]
brfss1315.shape
```

```python
# remove highBP=1, highchol=1,smoker=1,both mental and physical health problem days >25 ,but diabetes_012 != 2
brfss1315 = brfss1315[~((brfss1315['HighBP'] == 1) & (brfss1315['HighChol'] == 1) & (brfss1315['Smoker'] == 1) & (brfss1315['MentHlth'] >= 25) & (brfss1315['PhysHlth'] >= 25) & (brfss1315['Diabetes_012'] != 2))]
```

```python
# highBP=0, highchol=0,smoker=0,both mental and physical health problem days <= 5,but diabetes_012 != 0
brfss1315 = brfss1315[~((brfss1315['HighBP'] == 0) & (brfss1315['HighChol'] == 0) & (brfss1315['Smoker'] == 0) & (brfss1315['MentHlth'] <= 5) & (brfss1315['PhysHlth'] <= 5) & (brfss1315['Diabetes_012'] != 0))]
```

```python
brfss1315.shape
```

```python
is1 = brfss1315['Diabetes_012'] == 1
brfss1 = brfss1315[is1]
brfss1
```

```python
is0 = brfss1315['Diabetes_012'] == 0
brfss0 = brfss1315[is0]
brfss0
```

```python
is2 = brfss1315['Diabetes_012'] == 2
brfss2 = brfss1315[is2]
brfss2
```

```python
brfss0_rand1 = brfss0.take(np.random.permutation(len(brfss0))[:9226])
```

```python
brfss2_rand1 = brfss2.take(np.random.permutation(len(brfss2))[:9226])
```

```python
brfss_01 = brfss0_rand1.append(brfss1, ignore_index = True)
```

```python
brfss_012 = brfss_01.append(brfss2_rand1, ignore_index = True)
```
